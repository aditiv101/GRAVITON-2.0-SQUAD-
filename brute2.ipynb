```python
import numpy as np
import matplotlib.pyplot as plt
from itertools import permutations
from math import sqrt

# Function to calculate Euclidean distance between two cities
def euclidean_distance(city1, city2):
    return sqrt((city1[0] - city2[0]) ** 2 + (city1[1] - city2[1]) ** 2)

# Function to calculate the total distance of a tour
def calculate_distance(tour, cities):
    total_distance = 0
    for i in range(len(tour) - 1):
        total_distance += euclidean_distance(cities[tour[i]], cities[tour[i + 1]])
    total_distance += euclidean_distance(cities[tour[-1]], cities[tour[0]])
    return total_distance

# Brute force TSP solver
def tsp_brute_force(cities):
    number_of_cities = len(cities)
    min_distance = float('inf')
    best_tour = None

    for tour in permutations(range(number_of_cities)):
        current_distance = calculate_distance(tour, cities)
        if current_distance < min_distance:
            min_distance = current_distance
            best_tour = tour

    return best_tour, min_distance

# Visualize the tour
def visualize_tour(cities, tour):
    plt.figure(figsize=(10, 5))

    # Plot cities
    for city in cities:
        plt.plot(city[0], city[1], 'o', markersize=10)

    # Plot the tour
    for i in range(len(tour)):
        plt.arrow(cities[tour[i-1]][0], cities[tour[i-1]][1],
                  cities[tour[i]][0] - cities[tour[i-1]][0],
                  cities[tour[i]][1] - cities[tour[i-1]][1],
                  color='blue', length_includes_head=True, head_width=0.5, head_length=1)

    plt.title('Traveling Salesman Problem - Brute Force Solution')
    plt.xlabel('X Coordinate')
    plt.ylabel('Y Coordinate')
    plt.grid(True)
    plt.show()

# Example usage
if __name__ == "__main__":
    # Define cities as points (x, y)
    cities = [(0, 0), (10, 10), (10, 0), (0, 10)]

    best_tour, min_distance = tsp_brute_force(cities)
    print(f"Best tour: {best_tour}")
    print(f"Minimum distance: {min_distance}")

    visualize_tour(cities, best_tour)
```
